```python
file_name = 'input.txt'
numbers = []
operations = []
results = []


def read_input (file_name) :
    file = open(file_name, 'r').read().split("\n")
    lineNumber = len(file)
    numbers = []
    for i in range(lineNumber) :
        cleanedLine = file[i].split()
        #print('Line', i, ':', cleanedLine)
        if i == lineNumber - 1 :
            return numbers, cleanedLine
        else :
            numbers.append([int(x) for x in cleanedLine])


print('Read input file:', file_name)
numbers, operations = read_input(file_name)
#print('Numbers:', numbers)
#print('Operations:', operations)

for list in numbers :
    if len(list) != len(operations) :
        print('Error: Number of operations does not match number of values')
        exit(1)

operationPointer = 0
for operation in operations :
    if operation == '*' :
        result = 1
        for list in numbers :
            result = result * list[operationPointer]
    elif operation == '+' : 
        result = 0
        for list in numbers :
            result += list[operationPointer]
    results.append(result)
    operationPointer += 1

print('Results:', results)
print('Final Result:', sum(results))
```

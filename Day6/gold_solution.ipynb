```python
file_name = 'input.txt'
numbers = []
operations = []
results = []


def read_input (file_name) :
    file = open(file_name, 'r').read().split("\n")
    fileLenght = len(file)
    fileLenghtWithoutOperators = fileLenght - 1

    map = []
    mapOperation = [[] for x in range(fileLenghtWithoutOperators)]
    mapOperator = []
    indexLastNumber = 0

    for line in file :
        map.append(list(line))

    mapLenght = len(map[0])
    lastIndex = mapLenght - 1

    i = 0
    while indexLastNumber < mapLenght :
        startIndexNumber = indexLastNumber
        endIndexNumber = 0
        if (i >= lastIndex) :
            endIndexNumber = lastIndex + 1
            for l in range(fileLenghtWithoutOperators) :
                mapOperation[l].append(map[l][startIndexNumber:endIndexNumber])
            mapOperator.append(map[fileLenght-1][startIndexNumber])
            break
        if (map[0][i] == ' ' and map[1][i] == ' ' and map[2][i] == ' ' and map[3][i] == ' ') :
            endIndexNumber = i
            for l in range(fileLenghtWithoutOperators) :
                mapOperation[l].append(map[l][startIndexNumber:endIndexNumber])
            mapOperator.append(map[fileLenght-1][startIndexNumber])
            indexLastNumber = i + 1
        i += 1
    return mapOperation, mapOperator


def get_numbers_to_operate (lines) :
    numbers = []
    for e in range(len(lines[0])) :
        number = ''
        for line in lines :
            if line[e] != ' ' :
                number += line[e]
        if number == '' :
            print('Warning: element missing')
            print('lines', lines)
        else :
            numbers.append(int(number))
    return numbers


print('Read input file:', file_name)
numbers, operations = read_input(file_name)
print('Numbers:')
for number in numbers :
    print(number)
print('Operations:', operations)

for map in numbers :
    if len(map) != len(operations) :
        print('Error: Number of operations does not match number of values')
        exit(1)

operationPointer = 0
for operation in operations :
    numbersToCompute = get_numbers_to_operate([numbers[x][operationPointer] for x in range(len(numbers))])
    if operation == '*' :
        result = 1
        for number in numbersToCompute :
            result *= number
    elif operation == '+' : 
        result = sum(numbersToCompute)
    results.append(result)
    operationPointer += 1

print('Results:', results)
print('Final Result:', sum(results))
```

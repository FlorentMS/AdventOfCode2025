```python
file_name = 'input.txt'
max_dial = 99
position_possibilities = 100
current_position = 50
intrance_code = 0
instructions = []
print("Start at ", current_position)

def rotate (pointer, direction, tics) :
    global intrance_code
    if (direction == 'L') :
        tranlsation = - int(tics)
        # print("rotate left", tranlsation)
    elif (direction == 'R') :
        tranlsation = int(tics)
        # print("rotate right", tranlsation)
    else :
        print('parse error')
        return False
    
    pointer_result = (pointer + tranlsation) % position_possibilities
    intrance_code += abs((pointer + tranlsation) // position_possibilities)
    print(direction, tranlsation,'time on 0',abs((pointer + tranlsation) // position_possibilities))
    if(current_position == 0) :
        intrance_code -= 1 
    return pointer_result


file = open(file_name, 'r').read().split("\n\n")
for line in file:
    instructions = line.split('\n')


for rotation in instructions:
    direction = rotation[0]
    tics = rotation[1:]
    # print("Rotation  " + direction + " de " + tics)
    current_position = rotate (current_position, direction, tics)
    print('position', current_position)

print("Final position", current_position)
print("Code :", intrance_code)
 
```

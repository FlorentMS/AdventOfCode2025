```python
file_name = 'input.txt'
ranges = []
ids = []
isRange = True
freshIdCount = 0

def isFresh (id) :
    for range in ranges :
        if (id >= range[0] and id <= range[1]) :
            print('fresh ID', id, 'in range', range)
            return True
    return False


file = open(file_name, 'r').read().split("\n")
for line in file:
    if (isRange) :
        if (line == '') :
            isRange = False
        else :
            ranges.append([int(x) for x in line.split('-')])
    else :
        ids.append(int(line))


print('Start')
#print('ranges', ranges)
print('ids', ids)

for id in ids:
    if (isFresh(id)) :
        freshIdCount += 1

print('Fresh ids', freshIdCount)
```

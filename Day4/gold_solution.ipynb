```python
file_name = 'input.txt'
ranges = []
boundaries = []
globalRanges = []
freshIdCount = 0
minBoundary = 0
maxBoundary = 0

def is_fresh (id) :
    for r in ranges :
        if (id >= r[0] and id <= r[1]) :
            print('fresh ID', id, 'in range', r)
            return True
    return False

def insertion_sort(array, left=0, right=None):
    if right is None:
        right = len(array) - 1

    for i in range(left + 1, right + 1):
        key_item = array[i]
        j = i - 1
        while j >= left and array[j][0] > key_item[0]:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = key_item
    return array

file = open(file_name, 'r').read().split("\n")
for line in file:
    if (line == '') :
        break # Do not compute unuseful data
    else :
        gap = [int(x) for x in line.split('-')]

        ranges.append(gap)
        boundaries.append(gap[0])
        boundaries.append(gap[1])

        if (minBoundary == 0 or minBoundary > ranges[-1][0]) :
            minBoundary = ranges[-1][0]
        if (maxBoundary < ranges[-1][1]) :
            maxBoundary = ranges[-1][1]

print('Start')
print('min', minBoundary, ', max', maxBoundary)
print('diff', maxBoundary - minBoundary)
sortRanges= insertion_sort(ranges)

globalRanges.append(sortRanges[0])
for gap in sortRanges :
    if (gap[0] > globalRanges[-1][1]) :
        globalRanges.append(gap)
    elif (gap[1] > globalRanges[-1][1]) :
        globalRanges[-1][1] = gap[1]

for gap in globalRanges:
    gapLenght = (gap[1] - gap[0]) + 1
    freshIdCount += gapLenght

print('Fresh ids', freshIdCount)
```

```python
file_name = 'input.txt'
ranges = []
boundaries = []
globalRanges = []
freshIdCount = 0
minBoundary = 0
maxBoundary = 0

def insertion_sort(array, left=0, right=None):
    if right is None:
        right = len(array) - 1

    for i in range(left + 1, right + 1):
        key_item = array[i]
        j = i - 1
        while j >= left and array[j][0] > key_item[0]:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = key_item
    return array



file = open(file_name, 'r').read().split("\n")
for line in file:
    if (line == '') :
        break # Do not compute unuseful data
    else :
        gap = [int(x) for x in line.split('-')]
        ranges.append(gap)
        boundaries.append(gap[0])
        boundaries.append(gap[1])

print('Start')
sortRanges= insertion_sort(ranges) # Sort first range value to have simple merging rules

globalRanges.append(sortRanges[0])
# Merge ranges
for gap in sortRanges :
    if (gap[0] > globalRanges[-1][1]) :
        globalRanges.append(gap)
    elif (gap[1] > globalRanges[-1][1]) :
        globalRanges[-1][1] = gap[1]

# Compute fresh ids count on merged ranges
for gap in globalRanges:
    gapLenght = (gap[1] - gap[0]) + 1
    freshIdCount += gapLenght

print('Fresh ids', freshIdCount)
```

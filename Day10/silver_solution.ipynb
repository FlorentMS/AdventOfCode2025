```python
file_name = 'small_input.txt'

lightIndicators = []
lightIndicatorsOn = '#'
lightIndicatorsOff =  '.'

buttonsWiring = []

joltageRequirements = []

def read_input (file_name, detailsLog = False) :
    print('Read input file:', file_name)
    file = open(file_name, 'r').read().split("\n")
    formatedLines = []
    if (detailsLog) :
        print('Elements :')
    for line in file :
        elements = line.split(' ')
        if (detailsLog) :
            print(elements)
        formatedLines.append([[],[],[]])

        ## Format line data ##
        for i in range(len(elements)) :
            cleanedElem = elements[i][1:-1] # Remove first and last character
            #print('cleanedElem', i, ':', cleanedElem)
            if (i == 0) :
                # light indicators
                formatedLines[-1][0] = bool_format_light(cleanedElem)
            elif (i == len(elements) - 1) :
                # joltage requirements
                formatedLines[-1][2] = [int(x) for x in cleanedElem.split(',')]
            else :
                # buttons wiring
                formatedLines[-1][1].append([int(x) for x in cleanedElem.split(',')])

    if (detailsLog) :
        print('Formated lines :')
        for line in formatedLines :
            print(line)
    return formatedLines

def use_button(boolLightMap, buttons) :
    for index in buttons :
        boolLightMap[index] = not(boolLightMap[index])
    return boolLightMap

def pretty_format_light(boolLightMap, withBorder = False) :
    prettyMap = ''
    for bool in boolLightMap :
        if (bool) :
            prettyMap += lightIndicatorsOn
        else :
            prettyMap += lightIndicatorsOff
    if (withBorder) :
        return '[' + prettyMap + ']'
    else :
        return prettyMap
    
def bool_format_light(prettyLightMap, withBorder = False) :
    if (withBorder) :
        prettyLightMap = prettyLightMap[1:-1]
    return [x == lightIndicatorsOn for x in prettyLightMap]


machinesManualLines = read_input(file_name)

for instruction in machinesManualLines :
    lightGoal = instruction[0]
    print('Goal')
    print(pretty_format_light(lightGoal, True))
    lightIndicators = [False for x in lightGoal]
    buttons = instruction[1]

    # Resolve_puzzle -> try on each button, try 2 combination on each, try 3 combination on each
    lightIndicators = use_button(lightIndicators, buttons[4])
    print(pretty_format_light(lightIndicators, True))
    lightIndicators = use_button(lightIndicators, buttons[5])
    if (lightGoal == lightIndicators) :
        print('Goal reached !')
        print(pretty_format_light(lightIndicators, True))
        break




```

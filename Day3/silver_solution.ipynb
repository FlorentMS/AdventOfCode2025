```python
file_name = 'input.txt'
batteries = []
voltage = 0

def find_highest (list, max = 9) :
    maxList = 0
    index_maxList = 0
    for i in range(len(list)) :
        if (list[i] > maxList) :
            if (list[i] <= max) :
                maxList = list[i]
                index_maxList = i
    return maxList, index_maxList

def find_two_highest (bank) :
    sec_highest = 0
    i = 9
    while (sec_highest == 0) and (i > 0) :
        first_highest, first_index = find_highest(bank, i)
        # print('first (', first_highest, first_index, ')')
        sec_highest, sec_index = find_highest(bank[first_index+1:])
        sec_index += first_index+1
        # print('second (', sec_highest, sec_index, ')')
        i -= 1
    return first_highest, first_index, sec_highest, sec_index



file = open(file_name, 'r').read().split("\n")
for line in file:
    batteries.append([int(i) for i in list(line)])


print("Start")

for bank in batteries:
    # print('bank', bank)
    # print(find_two_highest (bank))
    first_voltage, firstVoltage_index, sec_voltage, secVoltage_index = find_two_highest (bank)
    bankVoltage = int(str(bank[firstVoltage_index]) + str(bank[secVoltage_index]))
    print('voltage bank', bankVoltage)
    voltage += bankVoltage

print(voltage)
```
